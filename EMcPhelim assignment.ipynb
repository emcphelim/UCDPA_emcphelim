```python
df1=pandas.read_csv("amazon_prime_titles.csv.zip")
```

```python
print(df1)
```

```python
import pandas as pd
```

```python
df2=pandas.read_csv("netflix_titles.csv.zip")
```

```python
print(df2)
```

```python
df3=pd.merge(df1,df2,on="title")
df3
```

```python
df4=pd.merge(df1,df2,on="title",how="inner")
df4
```

```python
df5=pd.merge(df1,df2,on="title",how="outer")
df5
```

```python
df6=pd.merge(df1,df2,on="title",how="left")
df6
```

```python
df7=pd.merge(df1,df2,on="title",how="left",indicator=True)
df7
```

```python
df8=pd.merge(df1,df2,on="title",suffixes=("_left","_right"))
df8
```

```python
data=pandas.read_csv("seattle-weather.csv")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots()
plt.plot(data["temp_max"],data["precipitation"],marker="o")
plt.xlabel("temp_max")
plt.ylabel("precipitation")
plt.title("temp_max Vs precipation")
plt.show()
```

```python
fig,ax=plt.subplots()
plt.scatter(data["temp_min"],data["wind"],marker="v")
plt.xlabel("temp_min")
plt.ylabel("wind")
plt.title("temp_min Vs wind")
plt.show()
```

```python
data["temp_min"].corr(data["wind"])
```

```python
import numpy as np
```

```python
min_array=data[["temp_min"]].to_numpy()
```

```python
wind_array=data[["wind"]].to_numpy()
```

```python
fig,ax=plt.subplots()
plt.bar(data["weather"],data["precipitation"])
plt.xlabel("weather")
plt.ylabel("precipitation")
plt.title("weather Vs precipitation")
plt.show()
```
